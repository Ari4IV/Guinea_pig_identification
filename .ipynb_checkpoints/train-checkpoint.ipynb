# 先導入所有需要的套件

```python
import tensorflow as tf # 導入 TensorFlow 函式庫
import numpy as np  # 匯入NumPy，是Python的一個數學函式庫，常用於資料操作
import pandas as pd  # 匯入Pandas，是Python的一個資料分析函式庫
import matplotlib.pyplot as plt  # 從Matplotlib套件匯入pyplot模組，用於繪製圖表和視覺化
import os  # 匯入os模組，提供了很多操作系統的功能
import seaborn as sns  # 匯入Seaborn，是基於Matplotlib的資料視覺化函式庫
from tensorflow.keras.utils import to_categorical  # 從TensorFlow匯入一個工具，用於將標籤轉換為one-hot編碼
from tensorflow.keras.preprocessing.image import load_img, img_to_array  # 從TensorFlow匯入影像處理工具
from sklearn.model_selection import train_test_split  # 從scikit-learn匯入工具，用於分割資料集
from tensorflow.keras.models import Sequential  # 從TensorFlow匯入Sequential模型類型
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D  # 從TensorFlow匯入神經網路層
from tensorflow.keras.optimizers import SGD  # 從TensorFlow匯入隨機梯度下降優化器
from ipywidgets import interact_manual  # 匯入ipywidgets中的互動工具，常用於Jupyter Notebook的互動功能
# 從 scikit-learn 載入 confusion_matrix 函數來計算混淆矩陣
from sklearn.metrics import classification_report, confusion_matrix  # 從scikit-learn匯入評估工具
# 載入 seaborn 函式庫以視覺化資料
```

# 資料預處理

```python
data_dir = './images'  # 定義資料夾的路徑
breeds = ['Abyssinian', 'American', 'Silkie', 'Skinny']  # 定義品種列表
images = []  # 建立空的圖片列表
labels = []  # 建立空的標籤列表

for idx, breed in enumerate(breeds):  # 列舉品種列表並取得索引和品種名稱
    breed_dir = os.path.join(data_dir, breed)  # 獲得特定品種的目錄路徑
    for img_name in os.listdir(breed_dir):  # 迭代該品種目錄下的所有檔案名稱
        if not img_name.startswith('.'):  # 排除以'.'開頭的隱藏檔案
            img_path = os.path.join(breed_dir, img_name)  # 獲得圖片的完整路徑
            img = load_img(img_path, target_size=(128, 128))  # 讀取圖片並調整其大小為 128x128
            images.append(img_to_array(img))  # 將圖片轉為數組並添加到 images 列表中
            labels.append(idx)  # 添加品種索引到標籤列表中

# 轉換成 numpy array
x_data = np.array(images, dtype='float32') / 255.  # 將圖片列表轉為 numpy 數組並進行正規化（0到1之間）
y_data = to_categorical(labels, num_classes=len(breeds))  # 將標籤列表轉為 one-hot 編碼格式
```

# 切分資料

```python
# 使用 train_test_split 函數分割資料集
# x_data 和 y_data 是輸入和對應的標籤
# test_size=0.2 表示測試集占總資料的 20%
# random_state=42 是隨機種子，確保每次分割都得到相同的結果
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

# 隨機查看一筆資料的長相

```python
print(x_train.shape)  # 打印 x_train 的形狀，通常是 (圖片數量, 高度, 寬度, 通道數)

# 從 [0, x_train.shape[0] - 1] 隨機選擇一個整數，這將用於從 x_train 中隨機選擇一張圖片
n = np.random.randint(x_train.shape[0])

# 使用 matplotlib 的 imshow 函數顯示選擇的圖片
# 由於圖片是灰階格式（假設是這種格式），所以使用 'Greys' 色彩映射
plt.imshow(x_train[n], cmap='Greys')

# 獲取 y_train 中對應圖片的標籤（one-hot 編碼格式），並找到其最大值的索引
# 使用 breeds 列表獲取該索引對應的品種名稱，並設置為圖片的標題
plt.title(breeds[np.argmax(y_train[n])])

# 顯示圖片及其標題
plt.show()
```

# 打造神經網路

```python
# 創建一個順序模型
model = Sequential([
    # 第一層是一個卷積層，有 32 個過濾器，每個過濾器大小為 (3, 3)。
    # 使用 'same' 填充，這意味著輸入圖像的大小在這一層不會改變。
    # 使用 ReLU (Rectified Linear Unit) 作為激活函數。
    # 並指定輸入圖片的形狀為 (128, 128, 3)，這意味著圖片是 128x128 且是 RGB 三通道。
    Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(128, 128, 3)),
    
    # 最大池化層，用於下採樣。使用 (2, 2) 的窗口大小，這將圖像的寬度和高度都減半。
    MaxPooling2D(2, 2),
    
    # 第二個卷積層，有 64 個過濾器，每個過濾器的大小為 (3, 3)。還是使用 'same' 填充和 ReLU 激活函數。
    Conv2D(64, (3, 3), padding='same', activation='relu'),
    
    # 又一個最大池化層，用於進一步下採樣。
    MaxPooling2D(2, 2),
    
    # 扁平化層，用於將 2D 特徵圖轉化為 1D 向量，以供全連接層使用。
    Flatten(),
    
    # 全連接層，有 128 個神經元並使用 ReLU 激活函數。
    Dense(128, activation='relu'),
    
    # 最後的全連接層，用於輸出預測。神經元的數量等於品種的數量。
    # 使用 softmax 激活函數，這確保輸出是一組機率分數，其中每一個分數對應一個品種。
    Dense(len(breeds), activation='softmax')
])
```

# 組裝神經網路

```python
# 編譯模型
model.compile(
    # 使用 'categorical_crossentropy' 作為損失函數。
    # 這是多分類任務的常見選擇，適用於標籤是使用 one-hot 編碼的情況。
    loss='categorical_crossentropy',
    
    # 使用隨機梯度下降 (SGD) 作為優化器。
    # 由於在M1/M2 Mac上，使用v2.11+的tf.keras.optimizers.SGD會運行緩慢，
    # 所以我們選擇使用過去版本的SGD優化器。
    # 同時設置學習率 (learning_rate) 為 0.01。
    optimizer = tf.keras.optimizers.legacy.SGD(learning_rate=0.01),
    
    # 使用 'accuracy' 作為評估指標。
    # 這會在訓練過程中持續報告模型的準確度。
    metrics=['accuracy']
)
```

# 查看模型結構

```python
# 使用 model.summary() 會顯示模型的架構摘要，包括每一層的名稱、輸出形狀、參數數量等。
model.summary()
```

# 訓練

```python
# 訓練模型
history = model.fit(
    # 訓練資料
    x_train, y_train, 
    
    # 訓練35個循環（epochs）
    epochs=35,
    
    # 驗證資料
    # 每完成一次訓練循環後，模型都會在此資料上進行評估，但不會根據這些資料進行訓練。
    validation_data=(x_test, y_test)
)
```

# 預測

```python
# 使用模型對測試資料進行預測
predictions = model.predict(x_test)
```

# 評估正確率

```python
# 使用 np.argmax 在 predictions 的 axis=1 上取得最大值的索引。
# 這些索引代表模型預測的類別。
predicted_classes = np.argmax(predictions, axis=1)

# 使用 np.argmax 在 y_test 的 axis=1 上取得最大值的索引。
# 這些索引代表真實的類別。
true_classes = np.argmax(y_test, axis=1)

# 比較 predicted_classes 和 true_classes 來判斷預測是否正確。
# 然後計算正確預測的平均值來獲得準確度。
accuracy = np.mean(predicted_classes == true_classes)

# 打印準確度，並保留兩位小數。
print(f"Accuracy: {accuracy:.2f}")
```

# 儲存模型

```python
# 將模型儲存到指定的.h5文件中
model.save('./model/guinea_pig_model.keras')
```

```python
# 計算混淆矩陣
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # 從預測的機率分佈中取得最大機率的索引作為預測類別
y_true = np.argmax(y_test, axis=1)  # 從真實的 one-hot 編碼中取得類別索引
confusion = confusion_matrix(y_true, y_pred_classes)

# 計算敏感度（True Positive Rate）
# 敏感度 = TP / (TP + FN)
TP = confusion.diagonal()  # 取得混淆矩陣的對角線元素，即真正例（True Positives）
FN = confusion.sum(axis=1) - TP  # 計算假負例（False Negatives）
sensitivity = TP / (TP + FN)  # 計算敏感度

# 創建敏感度柱狀圖
plt.figure(figsize=(12, 6))
plt.bar(breeds, sensitivity)  # 繪製柱狀圖
plt.title('Sensitivity for Each Class') #各類別的敏感度
plt.xlabel('Class') # 類別
plt.ylabel('Sensitivity') # 敏感度
plt.xticks(rotation=45)  # 旋轉 x 軸的標籤，使其更易讀
plt.grid(axis='y')  # 在 y 軸上添加網格線

plt.show()
```

```python
# 計算混淆矩陣
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # 從預測的機率分佈中取得最大機率的索引作為預測類別
y_true = np.argmax(y_test, axis=1)  # 從真實的 one-hot 編碼中取得類別索引
confusion = confusion_matrix(y_true, y_pred_classes)

# 計算特異性（Specificity 或 True Negative Rate）
# 特異性 = TN / (TN + FP)
TN = confusion.sum() - confusion.sum(axis=0) - confusion.sum(axis=1) + confusion.diagonal()  # 計算真負例（True Negatives）
FP = confusion.sum(axis=1) - confusion.diagonal()  # 計算假正例（False Positives）
specificity = TN / (TN + FP)  # 計算特異性

# 創建特異性柱狀圖
plt.figure(figsize=(12, 6))
plt.bar(breeds, specificity)  # 繪製柱狀圖
plt.title('Specificity for Each Class') # 各類別的特異性
plt.xlabel('Class') # 類別
plt.ylabel('Specificity') # 特異性
plt.xticks(rotation=45)  # 旋轉 x 軸的標籤，使其更易讀
plt.grid(axis='y')  # 在 y 軸上添加網格線

plt.show()
```

```python
# 計算混淆矩陣
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # 從預測的機率分佈中取得最大機率的索引作為預測類別
y_true = np.argmax(y_test, axis=1)  # 從真實的 one-hot 編碼中取得類別索引
confusion = confusion_matrix(y_true, y_pred_classes)

# 計算準確度、精確度、召回率和 F1 分數
accuracy = np.mean(y_pred_classes == y_true)  # 計算準確度
report = classification_report(y_true, y_pred_classes, target_names=breeds, output_dict=True)  # 取得分類報告
precision = [report[breed]['precision'] for breed in breeds]  # 取得各類別的精確度
recall = [report[breed]['recall'] for breed in breeds]  # 取得各類別的召回率
f1_score = [report[breed]['f1-score'] for breed in breeds]  # 取得各類別的 F1 分數

# 創建性能度量柱狀圖
plt.figure(figsize=(12, 6))
plt.bar(breeds, precision, label='Precision') # 精確度
plt.bar(breeds, recall, label='Recall', alpha=0.5) # 召回率
plt.bar(breeds, f1_score, label='F1 Score', alpha=0.5) # F1 分數
plt.title('Performance Metrics for Each Class') # 各類別的性能指標
plt.xlabel('Class') # 類別
plt.ylabel('Score') # 得分
plt.xticks(rotation=45)  # 旋轉 x 軸的標籤，使其更易讀
plt.legend()  # 添加圖例
plt.grid(axis='y')  # 在 y 軸上添加網格線

print(f"準確度: {accuracy:.2f}")
print("分類報告:\n", classification_report(y_true, y_pred_classes, target_names=breeds))

plt.show()
```
